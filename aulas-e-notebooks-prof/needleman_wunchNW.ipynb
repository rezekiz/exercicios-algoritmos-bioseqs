```python
def formatar(x):
  if type(x) is int:
    return f"{x:>3d}"
  else:
    return f"{x:>3}"

def print_matrix(S1, S2, M):
  print(*list(" " + S1), sep = '   ')
  for x2, L in zip(S2, M):
    LL = [formatar(x) for x in L]
    print(x2, *LL)
  print()

def score_subst(x1, x2, g):
  if '-' in x1 + x2: return g
  if x1 == x2:
    return 2
  else:
    return -1

def NW(S1, S2, score_space, score_subst):
  """
  S1: primeira string
  S2: segunda string
  score_subst: função que recebe dois caracteres e devolve o score da substituição de um pelo outro
  """
  S2 = '-' + S2
  S1 = '-' + S1

  ncols = len(S1)
  nlins = len(S2)

  # Inicializar toda a matriz a zeros
  score = [[0 for _ in range(ncols)] for _ in range(nlins)]

  trace = [[0 for _ in range(ncols)] for _ in range(nlins)]

  # Inicializar a primeira linha
  score[0] = [C * score_space for C, _ in enumerate(S1)]
  trace[0] = [0 if C == 0 else 'E' for C, _ in enumerate(S1)]


  # Inicializar a primeira coluna
  for L, _ in enumerate(S2):
    score[L][0] = L * score_space
    trace[L][0] = 0 if L == 0 else 'C'


  for L, (X2, linha) in enumerate(zip(S2, score)):
    for C, (X1, V) in enumerate(zip(S1, linha)):
      if L > 0 and C > 0:
        diag = score[L - 1][C - 1] + score_subst(X1, X2, score_space)
        esq  = score[L    ][C - 1] + score_space
        cima = score[L - 1][C    ] + score_space


        escolhas = [diag, esq, cima]
        direcoes = "DEC"

        valor = max(*escolhas)

        trace[L][C] = direcoes[escolhas.index(valor)]

        score[L][C] = valor


  print_matrix(S1, S2, score)
  print_matrix(S1, S2, trace)

  return score[-1][-1], reconstroi(S1, S2, trace)

def reconstroi(S1, S2, trace):
  A1 = ''
  A2 = ''

  L = len(S2) - 1
  C = len(S1) - 1


  while trace[L][C] != 0:
    print(L, C, S2[L], S1[C], trace[L][C])
    if trace[L][C] == 'D':
      A1 = S1[C] + A1
      A2 = S2[L] + A2
      L -= 1
      C -= 1
    elif trace[L][C] == 'C':
      A1 = '-'   + A1
      A2 = S2[L] + A2
      L -= 1
    elif trace[L][C] == 'E':
      A1 = S1[C] + A1
      A2 = '-' + A2
      C -= 1

  print(A1, A2, sep = "\n")
  return (A1, A2)

NW("ACT", "ACCT", -4, score_subst)
```
